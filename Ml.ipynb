```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.feature_selection import SelectFromModel
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier , GradientBoostingClassifier , RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import  cross_validate , StratifiedKFold , GridSearchCV , RandomizedSearchCV, train_test_split, cross_val_predict
from sklearn.preprocessing import StandardScaler , PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder , LabelEncoder
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from imblearn.pipeline import make_pipeline
from sklearn.metrics import recall_score , f1_score , accuracy_score , confusion_matrix ,classification_report, make_scorer
from category_encoders import BinaryEncoder
```

```python
models = [
    LogisticRegression(),
    KNeighborsClassifier(),
    GaussianNB(),
    SVC(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    XGBClassifier()
]
```

```python
transformer = ColumnTransformer(transformers=[
    ('ohe' ,OneHotEncoder(sparse_output=False , drop = 'first') , ['Geography' , 'Gender','Card Type']),
    
    
] , remainder = 'passthrough')

transformer
```

```python
x,y=df.drop('Exited', axis=1 ), df['Exited'] 
```

```python
df['Exited'].value_counts()
```

```python
pl=make_pipeline(transformer , StandardScaler(),SMOTE(sampling_strategy={1:3038}),SelectFromModel(estimator=RandomForestClassifier()))
```

```python
pl
```

```python
pl = make_pipeline(
    transformer,
    StandardScaler(),
    SMOTE(sampling_strategy={1: 3038}),
    SelectFromModel(estimator=RandomForestClassifier(random_state=42)),
    DecisionTreeClassifier()  
)


pl.fit(x, y)


feature_names = transformer.get_feature_names_out()
selected_mask = pl.named_steps['selectfrommodel'].get_support()
selected_features = feature_names[selected_mask]
print('important features')
for col in selected_features:
    print(col)
```

```python
for model in models:
    pl =make_pipeline(transformer,StandardScaler(),SMOTE(sampling_strategy={1: 3038}),SelectFromModel(estimator=RandomForestClassifier()),model)

    cv = cross_validate(
        estimator=pl, 
        X=x,
        y=y,
        cv=StratifiedKFold(n_splits=5),
        scoring=['accuracy', 'recall_macro', 'f1_macro','precision_macro'],
        return_train_score=True
    ) 


    print(model)
    print(f'Avg Train Score :{cv['train_accuracy'].mean()}')
    print(f'Avg Test Score : {cv['test_accuracy'].mean()}')
    print(40*"*")
    print(f"Avg Recall      : {cv['train_recall_macro'].mean():.4f}")
    print(f"Avg Recall      : {cv['test_recall_macro'].mean():.4f}")
    print(40*"*")
    print(f"Avg F1 Score    : {cv['train_f1_macro'].mean():.4f}")
    print(f"Avg F1 Score    : {cv['test_f1_macro'].mean():.4f}")
    print(40*"*")
    print(f"Avg Precision Train: {cv['train_precision_macro'].mean():.4f}")
    print(f"Avg Precision Test : {cv['test_precision_macro'].mean():.4f}")
    print(40*"*")
    x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.2, random_state=42)
    pl.fit(x_train , y_train)
    y_train_prd = pl.predict(x_train)
    y_test_prd = pl.predict(x_test)
    print("ConfusionMatrix")
    print("📊 Confusion Matrix - Test:")
    print(confusion_matrix(y_test, y_test_prd))
    print("\nClassification Report - Test:")
    print(classification_report(y_test, y_test_prd))
```

```python
dt = make_pipeline(transformer,SMOTE(sampling_strategy={1: 3038}),SelectFromModel(estimator=RandomForestClassifier()) , DecisionTreeClassifier())
```

```python
dt
```

```python
dt_param = [
    {
        'decisiontreeclassifier__criterion' : ["gini", "entropy", "log_loss"],
        'decisiontreeclassifier__splitter' : ["best", "random"],
        'decisiontreeclassifier__max_depth' : [2,3,4,5],
        'decisiontreeclassifier__min_samples_split' : [2,3,4,5,]
    }
]
```

```python
> ℹ️ **Note**:  
Due to the large size of the parameter space and the high accuracy of the model (≈ 99%),  used **RandomizedSearchCV** instead of **GridSearchCV**.  
This choice significantly reduced computation time while still providing competitive performance.
```

```python
dt_srch = RandomizedSearchCV(
    estimator=dt,
    param_distributions=dt_param,
    n_iter=10,  
    scoring='recall_macro',
    cv=StratifiedKFold(n_splits=5),
    return_train_score=True,
    random_state=42,   
)
```

```python
dt_srch.fit(x , y)
```

```python
dt_srch.best_estimator_
```
